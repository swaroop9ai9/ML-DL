4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst

```python
# Initially lets convert the 'GoalsFor' and 'GoalsAgainst' in to integers 
laliga['GoalsFor']=laliga['GoalsFor'].astype(str).astype(int)
```

```python
laliga['GoalsAgainst']=laliga['GoalsAgainst'].astype(str).astype(int)
```

```python
#Defining the function 'Goal_diff_count' which returns a Series DataFrame with the 'Team' and 'Goal_diff' columns.
def Goal_diff_count(df):
    return pd.Series([df['Team'],int(df['GoalsFor']-df['GoalsAgainst'])],index=['Team','Goal_diff'])
```

```python
#Lets Save the Series Output in a variable
goal_count=laliga.apply(Goal_diff_count,axis=1)
```

```python
#Lets Verify the 'goal_count'
goal_count
```

```python
# We can print the Team with highest difference using the 'Goal_diff_count' Function to show the max value:
goal_count[goal_count['Goal_diff']==max(laliga.apply(Goal_diff_count,axis=1)['Goal_diff'])]
```

```python
# Similarly, We can print the Team with least difference using the 'Goal_diff_count' Function to show the min value:
goal_count[goal_count['Goal_diff']==min(laliga.apply(Goal_diff_count,axis=1)['Goal_diff'])]
```

5. Create a new column with name “Winning Percent” and append it to the data set
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
# Initially lets convert the 'GamesWon' & 'GamesPlayed' in to integers in the laliga datframe
laliga['GamesWon']=laliga['GamesWon'].astype(str).astype(int)
```

```python
laliga['GamesPlayed']=laliga['GamesPlayed'].astype(str).astype(int)
```

```python
# New Column with name 'Winning Percent' where Percentage of Winning =(GamesWon/GamesPlayed)*100
WinningPercent=((laliga['GamesWon']/laliga['GamesPlayed'])*100).round(2)
```

```python
#Appending the columns to the dataframe with column name 'Winning percent'
laliga['Winning Percent']=WinningPercent
```

```python
# Filling all NaN values with 0 percent
laliga['Winning Percent'].fillna(0,inplace=True)
```

```python
#Printing Top 5 teams which have Highest winning percentage
laliga.nlargest(5,'Winning Percent').loc[:,['Team','Winning Percent']]
```
